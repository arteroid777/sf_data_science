2. Напишите запрос, который для каждого региона выведет количество работодателей и вакансий в нём.
Среди регионов, в которых нет вакансий, найдите тот, в котором наибольшее количество работодателей.


```python
# Запрос для статистики по регионам с использованием IS NULL для оптимизации
query_5_2_fast = '''
SELECT 
    a.name AS region_name,
    COUNT(DISTINCT e.id) AS employers_count,
    COUNT(DISTINCT v.id) AS vacancies_count
FROM public.areas a
LEFT JOIN public.employers e ON a.id = e.area
LEFT JOIN public.vacancies v ON a.id = v.area_id
GROUP BY a.id, a.name
ORDER BY employers_count DESC, vacancies_count DESC;
'''

# Отдельный запрос для быстрого поиска региона без вакансий с наибольшим количеством работодателей
query_no_vacancies_fast = '''
SELECT 
    a.name AS region_name,
    COUNT(DISTINCT e.id) AS employers_count
FROM public.areas a
LEFT JOIN public.employers e ON a.id = e.area
WHERE NOT EXISTS (
    SELECT 1 
    FROM public.vacancies v 
    WHERE v.area_id = a.id
)
GROUP BY a.id, a.name
ORDER BY employers_count DESC
LIMIT 1;
'''
```

```python
# Выполняем запросы
print("Получаем статистику по регионам...")
df_regions = pd.read_sql_query(query_5_2_fast, connection)
print("Статистика по регионам:")
print(df_regions.head(10))  # Покажем первые 10 регионов

print("\nИщем регион без вакансий с наибольшим количеством работодателей...")
df_no_vacancies = pd.read_sql_query(query_no_vacancies_fast, connection)

if not df_no_vacancies.empty:
    print(f"\nРегион без вакансий с наибольшим количеством работодателей:")
    print(f"Регион: {df_no_vacancies['region_name'].iloc[0]}")
    print(f"Количество работодателей: {df_no_vacancies['employers_count'].iloc[0]}")
else:
    print("\nНет регионов без вакансий")
```

7. Для компании «Яндекс» выведите список регионов-миллионников, в которых представлены вакансии компании, вместе с количеством вакансий в этих регионах. Также добавьте строку Total с общим количеством вакансий компании. Результат отсортируйте по возрастанию количества.

Список городов-милионников надо взять [отсюда](https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D1%80%D0%BE%D0%B4%D0%B0-%D0%BC%D0%B8%D0%BB%D0%BB%D0%B8%D0%BE%D0%BD%D0%B5%D1%80%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8). 

Если возникнут трудности с этим задание посмотрите материалы модуля  PYTHON-17. Как получать данные из веб-источников и API. 

```python
# код для получения списка городов-милионников
million_cities = [
    'Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург', 'Казань',
    'Красноярск', 'Нижний Новгород', 'Челябинск', 'Уфа', 'Краснодар',
    'Самара', 'Ростов-на-Дону', 'Омск', 'Воронеж', 'Пермь', 'Волгоград'
]

print("Список городов-миллионников России:")
for i, city in enumerate(million_cities, 1):
    print(f"{i}. {city}")
```

```python
# текст запроса
query_5_7 = '''
WITH yandex_million_cities AS (
    -- Вакансии Яндекса в городах-миллионниках
    SELECT 
        a.name AS region_name,
        COUNT(*) AS vacancy_count
    FROM public.vacancies v
    JOIN public.employers e ON v.employer_id = e.id
    JOIN public.areas a ON v.area_id = a.id
    WHERE e.name = 'Яндекс' 
      AND a.name IN ('Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург', 'Казань',
                    'Красноярск', 'Нижний Новгород', 'Челябинск', 'Уфа', 'Краснодар',
                    'Самара', 'Ростов-на-Дону', 'Омск', 'Воронеж', 'Пермь', 'Волгоград')
    GROUP BY a.id, a.name
    
    UNION ALL
    
    -- Общее количество вакансий Яндекса в городах-миллионниках
    SELECT 
        'Total' AS region_name,
        COUNT(*) AS vacancy_count
    FROM public.vacancies v
    JOIN public.employers e ON v.employer_id = e.id
    JOIN public.areas a ON v.area_id = a.id
    WHERE e.name = 'Яндекс'
      AND a.name IN ('Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург', 'Казань',
                    'Красноярск', 'Нижний Новгород', 'Челябинск', 'Уфа', 'Краснодар',
                    'Самара', 'Ростов-на-Дону', 'Омск', 'Воронеж', 'Пермь', 'Волгоград')
)
SELECT region_name, vacancy_count
FROM yandex_million_cities
ORDER BY 
    CASE WHEN region_name = 'Total' THEN 1 ELSE 0 END,
    vacancy_count DESC;
'''
```

```python
# результат запроса
print("Выполняем запрос...")
df_yandex_vacancies = pd.read_sql_query(query_5_7, connection)
print("\nРезультат запроса:")
print(df_yandex_vacancies)
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*). 

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа. 

```python
# текст запроса
query_avg_salary_3_6_years = '''
SELECT 
    AVG(
        (COALESCE(salary_from, salary_to) + COALESCE(salary_to, salary_from)) / 2
    ) AS avg_salary_3_to_6_years
FROM public.vacancies
WHERE 
    -- Условие: хотя бы одно из полей с зарплатой заполнено
    (salary_from IS NOT NULL OR salary_to IS NOT NULL)
    AND
    -- Условие: опыт работы от 3 до 6 лет
    experience = 'От 3 до 6 лет'
    AND
    -- Условия для вакансий Data Scientist
    (name ILIKE '%data scientist%' OR
     name ILIKE '%data science%' OR
     name ILIKE '%исследователь данных%' OR
     (name ILIKE '%ml%' AND name NOT ILIKE '%html%') OR
     name ILIKE '%machine learning%' OR
     name ILIKE '%машинн%обучен%')
'''
```

```python
# результат запроса
df_avg_salary1 = pd.read_sql_query(query_avg_salary_3_6_years, connection)
print(df_avg_salary1)
```
